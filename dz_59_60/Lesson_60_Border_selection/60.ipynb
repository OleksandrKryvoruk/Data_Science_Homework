```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./Img/1050.jpg')
```

```python
img = cv2.resize(img, (1024, 768))
```

```python
cv2.imshow('', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
fig, _ = plt.subplots(figsize=(8, 10))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB) )
plt.axis('off')
plt.show()
```

### Sobel Filtering

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
kernel_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
kernel_y = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
```

```python
sobel_x = cv2.filter2D(gray, -1, kernel_x)
sobel_y = cv2.filter2D(gray, -1, kernel_y)
sobel = sobel_x + sobel_y
```

```python
# cv2.imshow('Original images', img)
cv2.imshow('SOBEL', sobel)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imwrite('./Img/sobel.jpg', sobel)
```

```python
fig, _ = plt.subplots(figsize=(8, 10))
plt.imshow(cv2.cvtColor(sobel, cv2.COLOR_BGR2RGB) )
plt.axis('off')
plt.show()
```

### Canny Filtering

```python
# Convert image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
canny = cv2.Canny(gray, 50, 150)
```

```python
# cv2.imshow('Original images', img)
cv2.imshow('CANNY', canny)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imwrite('./Img/canny.jpg', canny)
```

```python
fig, _ = plt.subplots(figsize=(8, 10))
plt.imshow(cv2.cvtColor(canny, cv2.COLOR_BGR2RGB) )
plt.axis('off')
plt.show()
```

### Roberts Filtering

```python
kernelx = np.array([[1, 0],
                    [0, -1]])
kernely = np.array([[0, 1],
                    [-1, 0]])

img_robertx = cv2.filter2D(img, -1, kernelx)
img_roberty = cv2.filter2D(img, -1, kernely)
roberts = cv2.addWeighted(img_robertx, 0.5, img_roberty, 0.5, 0)
```

```python
# roberts = cv2.convertScaleAbs(roberts, alpha=2, beta=2)
cv2.imshow('Robert`s', roberts)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imwrite('./Img/roberts.jpg', roberts)
```

```python
fig, _ = plt.subplots(figsize=(8, 10))
plt.imshow(cv2.cvtColor(roberts, cv2.COLOR_BGR2RGB) )
plt.axis('off')
plt.show()
```
